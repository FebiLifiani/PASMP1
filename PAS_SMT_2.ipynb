```python
import os
import random
from shutil import copyfile

# Direktori utama
base_dir = '/content/drive/MyDrive/MP1B/cnn2'

# Direktori data sumber (bahan gambar per kelas)
Ayam_dir = os.path.join(base_dir, 'bahan', 'Ayam')
Kuda_dir = os.path.join(base_dir, 'bahan', 'Kuda')
Sapi_dir = os.path.join(base_dir, 'bahan', 'Sapi')

# Direktori tujuan latih
train_Ayam = os.path.join(base_dir, 'latih', 'Ayam')
train_Kuda = os.path.join(base_dir, 'latih', 'Kuda')
train_Sapi = os.path.join(base_dir, 'latih', 'Sapi')

# Direktori tujuan validasi
validasi_Ayam = os.path.join(base_dir, 'validasi', 'Ayam')
validasi_Kuda = os.path.join(base_dir, 'validasi', 'Kuda')
validasi_Sapi = os.path.join(base_dir, 'validasi', 'Sapi')

# Buat semua folder tujuan jika belum ada
for path in [train_Ayam, train_Kuda, train_Sapi, validasi_Ayam, validasi_Kuda, validasi_Sapi]:
    os.makedirs(path, exist_ok=True)

# Cek jumlah data asli
print("Jumlah Data Tiap Kelas")
print("Ayam: ", len(os.listdir(Ayam_dir)))
print("Kuda: ", len(os.listdir(Kuda_dir)))
print("Sapi: ", len(os.listdir(Sapi_dir)))

# Fungsi split data
def train_val_split(source, train, val, train_ratio):
    files = os.listdir(source)
    total_size = len(files)
    train_size = int(total_size * train_ratio)

    randomized = random.sample(files, total_size)
    train_files = randomized[:train_size]
    val_files = randomized[train_size:]

    for i in train_files:
        i_file = os.path.join(source, i)
        destination = os.path.join(train, i)
        copyfile(i_file, destination)

    for i in val_files:
        i_file = os.path.join(source, i)
        destination = os.path.join(val, i)
        copyfile(i_file, destination)

# Rasio data latih
train_ratio = 0.7

# Jalankan pembagian untuk tiap kelas
train_val_split(Ayam_dir, train_Ayam, validasi_Ayam, train_ratio)
train_val_split(Kuda_dir, train_Kuda, validasi_Kuda, train_ratio)
train_val_split(Sapi_dir, train_Sapi, validasi_Sapi, train_ratio)
```

Pre- Processing

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

Target

```python
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 30,
    horizontal_flip = True,
    shear_range = 0.3,
    fill_mode = 'nearest',
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    zoom_range = 0.1,

)

val_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 30,
    horizontal_flip = True,
    shear_range = 0.3,
    fill_mode = 'nearest',
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    zoom_range = 0.1,
)
```

```python
train_generator = train_datagen.flow_from_directory(
    os.path.join(base_dir, 'latih'),  # Ganti dari base_dir ke latih
    target_size=(150, 150),
    batch_size=128,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    os.path.join(base_dir, 'validasi'),  # Ganti dari base_dir ke validasi
    target_size=(150, 150),
    batch_size=128,
    class_mode='categorical'
)
```

```python
print("Train Ayam:", len(os.listdir(train_Ayam)))
print("Train Kuda:", len(os.listdir(train_Kuda)))
print("Train Sapi:", len(os.listdir(train_Sapi)))
print("Val Ayam:", len(os.listdir(validasi_Ayam)))
print("Val Kuda:", len(os.listdir(validasi_Kuda)))
print("Val Sapi:", len(os.listdir(validasi_Sapi)))
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.9):
      print('\nAkurasi mencapai 99%')
      self.model.stop_traning = True

callback = myCallback()
```

Model CNN

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(200, activation='relu'),
    tf.keras.layers.Dropout(0.3, seed=112),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dropout(0.5, seed=112),
    tf.keras.layers.Dense(3, activation='softmax')  # 3 classes
])

model.summary()
```

Tingkat Akurasi dan loss model

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(),
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=5,
    epochs=50,
    validation_data=val_generator,
    validation_steps=1,
    verbose=1  # Optional: ganti ke 1 kalau mau lihat progress bar
)
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc= 'best')
plt.show()

plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend(loc= 'best')

plt.show()
```
